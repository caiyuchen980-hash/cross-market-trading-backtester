```python
# ======================== 股票期权综合分析工具 ========================
# 功能说明：
# 1. 自由输入股票代码，自动识别市场：
#    - A股：直接输入数字（如 600519、000858）→ 自动加后缀
#    - 港股：直接输入数字（如 700、941）→ 自动加 .HK
#    - 美股：直接输入英文（如 PLTR、AAPL、TSLA）
# 2. 绘制股价走势图（MA5、MA20均线）
# 3. 文字显示：5日线、20日线、支撑位、压力位、突破状态
# 4. 抓取期权数据（大单、资金、PCR、MaxPain、IV）
# 5. 自动生成期权交易策略
# 6. 绘制期权持仓分布图 + 波动率曲线
# 7. 五大策略回测（Buy&Hold / MACD / RSI / MA信封 / MA振荡器）
# ================================================================================

# 关闭Jupyter输出截断
from IPython.display import display, HTML
display(HTML("<style>.output_scroll {max-height: none !important;}</style>"))

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import warnings
warnings.filterwarnings("ignore")

# 解决中文显示
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

# ====================== 自动识别市场 ======================
def auto_code(ticker):
    s = str(ticker).strip().upper()
    if s.endswith((".SS", ".SZ", ".HK")):
        return s
    if s.isdigit():
        if len(s) <= 5:
            return s.zfill(5) + ".HK"
        elif s.startswith(("6", "8")):
            return s + ".SS"
        elif s.startswith(("0", "3")):
            return s + ".SZ"
    return s

# ==============================
# 工具函数（回测专用）
# ==============================
def to_list(s):
    if isinstance(s, pd.DataFrame):
        s = s.iloc[:, 0]
    return [float(x) for x in s]

def max_drawdown(equity_list):
    eq = np.array(equity_list)
    peak = np.maximum.accumulate(eq)
    drawdown = (eq - peak) / peak
    return round(drawdown.min() * 100, 2)

def calculate_win_rate(trades):
    if len(trades) < 2:
        return 100.0
    wins = 0
    for i in range(0, len(trades), 2):
        if i+1 >= len(trades): continue
        buy_price = trades[i][1]
        sell_price = trades[i+1][1]
        if sell_price > buy_price:
            wins += 1
    total = len(trades) // 2
    return round(wins / total * 100, 1) if total > 0 else 100.0

def support_resistance(close, window=20):
    short_supp = round(min(close[-window:]), 2)
    short_res = round(max(close[-window:]), 2)
    long_supp  = round(min(close[-120:]), 2)
    long_res   = round(max(close[-120:]), 2)
    return short_supp, short_res, long_supp, long_res

def trend_direction(close):
    ma50 = np.mean(close[-50:])
    ma200 = np.mean(close[-200:]) if len(close)>=200 else ma50
    last = close[-1]
    if last > ma50 and last > ma200:
        return "Bullish (Strong Up Trend)"
    elif last > ma50:
        return "Neutral-Bullish"
    elif last < ma50 and last < ma200:
        return "Bearish (Down Trend)"
    else:
        return "Sideways (Range Bound)"

def option_flow_estimate(close):
    chg = (close[-1] - close[-2]) / close[-2]
    vol = np.std(close[-20:])
    if chg > 0 and vol > 0.02:
        return "Call Flow Dominant (Bullish Activity)"
    elif chg < 0 and vol > 0.02:
        return "Put Flow Dominant (Bearish Activity)"
    else:
        return "Low Volume / Neutral Flow"

# ==============================
# 策略 1：Buy & Hold
# ==============================
def backtest_bh(df):
    close = to_list(df["Close"])
    capital = 10000
    pos = capital // close[0]
    capital -= pos * close[0]
    equity = [capital + pos * c for c in close]
    final = equity[-1]
    ret = round((final - 10000)/10000, 2)
    trades = [(df.index[0], close[0], "BUY"), (df.index[-1], close[-1], "SELL")]
    mdd = max_drawdown(equity)
    win_rate = calculate_win_rate(trades)
    # 修复：补全与其他策略一致的返回结构，表格不再为空
    return final, ret, trades, None, "BuyHold", equity, mdd, win_rate

# ==============================
# 策略 2：MACD
# ==============================
def backtest_macd(df):
    close = to_list(df["Close"])
    dates = df.index
    ema12 = to_list(df["Close"].ewm(span=12).mean())
    ema26 = to_list(df["Close"].ewm(span=26).mean())
    macd = [ema12[i]-ema26[i] for i in range(len(ema12))]
    sig = to_list(pd.Series(macd).ewm(span=9).mean())

    capital, pos, buy = 10000, 0, 0
    trades, equity = [], []
    for i in range(len(close)):
        c = close[i]
        equity.append(capital + pos*c if pos else capital)
        if i < 1: continue
        m, s, mp, sp = macd[i], sig[i], macd[i-1], sig[i-1]
        if pos == 0:
            if mp < sp and m > s:
                pos = capital//c
                capital -= pos*c
                buy = c
                trades.append((dates[i],c,"BUY"))
        else:
            if (mp>sp and m<s) or c<=buy*0.95 or c>=buy*1.15:
                capital += pos*c
                pos = 0
                trades.append((dates[i],c,"SELL"))
    if pos:
        capital += pos*close[-1]
        trades.append((dates[-1], close[-1], "SELL"))
    ret = round((capital-10000)/10000,2)
    mdd = max_drawdown(equity)
    win_rate = calculate_win_rate(trades)
    return capital, ret, trades, (macd, sig), "MACD", equity, mdd, win_rate

# ==============================
# ==============================
# ==============================
# 策略 3：RSI 
# ==============================
def backtest_rsi(df):
    close = to_list(df["Close"])
    dates = df.index
    
    delta = df["Close"].diff().dropna()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()
    rs = avg_gain / avg_loss
    rsi_series = 100 - (100 / (1 + rs))
    
    rsi_series = rsi_series.squeeze()
    rsi = [np.nan] * 14 + rsi_series.dropna().tolist()

    capital, pos, buy = 10000,0,0
    trades, equity = [], []
    for i in range(len(close)):
        c = close[i]
        equity.append(capital + pos*c if pos else capital)
        if i < 14 or pd.isna(rsi[i]):
            continue
        r = rsi[i]
        if pos == 0:
            if r < 30:
                pos = capital//c
                capital -= pos*c
                buy = c
                trades.append((dates[i],c,"BUY"))
        else:
            if r>70 or c<=buy*0.95:
                capital += pos*c
                pos=0
                trades.append((dates[i],c,"SELL"))
    if pos:
        capital += pos*close[-1]
        trades.append((dates[-1], close[-1], "SELL"))
    ret = round((capital-10000)/10000,2)
    mdd = max_drawdown(equity)
    win_rate = calculate_win_rate(trades)
    return capital, ret, trades, rsi, "RSI", equity, mdd, win_rate


# ==============================
# 策略 4：MA Envelope
# ==============================
def backtest_ma_envelope(df):
    close = to_list(df["Close"])
    dates = df.index
    ma20 = to_list(df["Close"].rolling(20).mean())
    upper = [ma20[i]*1.03 for i in range(len(ma20))]
    lower = [ma20[i]*0.97 for i in range(len(ma20))]

    capital, pos, buy = 10000,0,0
    trades, equity = [], []
    for i in range(len(close)):
        c = close[i]
        equity.append(capital + pos*c if pos else capital)
        if i < 20: continue
        if pos ==0:
            if c < lower[i]:
                pos = capital//c
                capital -= pos*c
                buy = c
                trades.append((dates[i],c,"BUY"))
        else:
            if c > upper[i] or c <= buy*0.95:
                capital += pos*c
                pos=0
                trades.append((dates[i],c,"SELL"))
    if pos:
        capital += pos*close[-1]
        trades.append((dates[-1], close[-1], "SELL"))
    ret = round((capital-10000)/10000,2)
    mdd = max_drawdown(equity)
    win_rate = calculate_win_rate(trades)
    return capital, ret, trades, (ma20, upper, lower), "MA Envelope", equity, mdd, win_rate

# ==============================
# 策略 5：MA Oscillator
# ==============================
def backtest_ma_osc(df):
    close = to_list(df["Close"])
    dates = df.index
    ma10 = to_list(df["Close"].rolling(10).mean())
    ma30 = to_list(df["Close"].rolling(30).mean())
    osc = [ma10[i]-ma30[i] for i in range(len(ma10))]

    capital, pos, buy = 10000,0,0
    trades, equity = [], []
    for i in range(len(close)):
        c = close[i]
        equity.append(capital + pos*c if pos else capital)
        if i < 30: continue
        os = osc[i]
        osp = osc[i-1]
        if pos ==0:
            if osp < 0 and os > 0:
                pos = capital//c
                capital -= pos*c
                buy = c
                trades.append((dates[i],c,"BUY"))
        else:
            if osp >0 and os <0 or c <= buy*0.95:
                capital += pos*c
                pos=0
                trades.append((dates[i],c,"SELL"))
    if pos:
        capital += pos*close[-1]
        trades.append((dates[-1], close[-1], "SELL"))
    ret = round((capital-10000)/10000,2)
    mdd = max_drawdown(equity)
    win_rate = calculate_win_rate(trades)
    return capital, ret, trades, osc, "MA Oscillator", equity, mdd, win_rate

# ==============================
# Bloomberg 风格绘图
# ==============================
def plot_bloomberg(df, name, trades, ind, ind_name, equity, ret, mdd, win_rate, sup_res_info, trend, flow):
    plt.style.use('dark_background')
    close = to_list(df["Close"])
    dates = df.index
    ss, sr, ls, lr = sup_res_info

    fig = plt.figure(figsize=(16,10))
    gs = GridSpec(3,5,figure=fig,width_ratios=[3,3,3,3,1])

    ax1 = fig.add_subplot(gs[0:2,0:4])
    ax1.plot(dates, close, color='white',lw=1.2)
    
    try:
        if trades is not None and isinstance(trades, list) and len(trades) > 0:
            for d,p,act in trades:
                if act=="BUY":
                    ax1.scatter(d,p,c='#2ecc71',marker='^',s=100,zorder=5)
                else:
                    ax1.scatter(d,p,c='#e74c3c',marker='v',s=100,zorder=5)
    except:
        pass
    
# ==============================
# Bloomberg 风格绘图 
# ==============================
def plot_bloomberg(df, name, trades, ind, ind_name, equity, ret, mdd, win_rate, sup_res_info, trend, flow):
    plt.style.use('dark_background')
    close = to_list(df["Close"])
    dates = df.index
    ss, sr, ls, lr = sup_res_info

    fig = plt.figure(figsize=(16,10))
    gs = GridSpec(3,5,figure=fig,width_ratios=[3,3,3,3,1])

    ax1 = fig.add_subplot(gs[0:2,0:4])
    ax1.plot(dates, close, color='white', lw=1.2, label='Price')
    
    try:
        if trades is not None and isinstance(trades, list) and len(trades) > 0:
            for d,p,act in trades:
                if act=="BUY":
                    ax1.scatter(d,p,c='#2ecc71',marker='^',s=120,zorder=5)
                else:
                    ax1.scatter(d,p,c='#e74c3c',marker='v',s=120,zorder=5)
    except:
        pass
    
    ax1.set_ylabel('Price USD', c='white')
    ax1.grid(alpha=0.3, color='gray')
    ax1.tick_params(colors='white')

    show_ax2 = True
    if ind is not None:
        if ind_name == "MA Envelope":
            ma20, upper, lower = ind
            ax1.plot(dates, ma20, c='#f1c40f', lw=1.8, label='MA20')
            ax1.plot(dates, upper, c='#e74c3c', lw=1.2, alpha=0.8, label='Upper Band')
            ax1.plot(dates, lower, c='#2ecc71', lw=1.2, alpha=0.8, label='Lower Band')
            ax1.fill_between(dates, lower, upper, color='cyan', alpha=0.06)
            ax1.legend(loc='upper left', facecolor='#222')
            show_ax2 = False

    if show_ax2 and ind is not None and ind_name != "BuyHold":
        ax2 = fig.add_subplot(gs[2,0:4])
        if ind_name=="MACD":
            ax2.plot(dates,ind[0],c='#9b59b6',label='MACD')
            ax2.plot(dates,ind[1],c='#f39c12',label='Signal')
        elif ind_name=="RSI":
            ax2.plot(dates, ind, c='#3498db',label='RSI')
            ax2.axhline(30,c='#2ecc71',ls='--')
            ax2.axhline(70,c='#e74c3c',ls='--')
            ax2.set_ylim(0,100)
        elif ind_name=="MA Oscillator":
            ax2.plot(dates,ind,c='#9b59b6',label='MA Osc')
            ax2.axhline(0,c='white',ls='--')
        ax2.legend(facecolor='#222',edgecolor='gray')
        ax2.set_ylabel('Indicator',c='white')
        ax2.grid(alpha=0.3,color='gray')
        ax2.tick_params(colors='white')
    else:
        if ind_name == "BuyHold":
            gs.update(hspace=0.05)
            ax1 = fig.add_subplot(gs[0:3,0:4])
            ax1.plot(dates, close, color='white',lw=1.2)
            try:
                if trades is not None and isinstance(trades, list) and len(trades) > 0:
                    for d,p,act in trades:
                        if act=="BUY":
                            ax1.scatter(d,p,c='#2ecc71',marker='^',s=100,zorder=5)
                        else:
                            ax1.scatter(d,p,c='#e74c3c',marker='v',s=100,zorder=5)
            except:
                pass
            ax1.set_ylabel('Price USD',c='white')
            ax1.grid(alpha=0.3,color='gray')
            ax1.tick_params(colors='white')

    ax3 = fig.add_subplot(gs[0:3,4])
    ax3.axis('off')
    txt = f"""
Strategy: {name}
------------------------
Trades: {len(trades)//2 if isinstance(trades,list) else 0}
Win Rate: {win_rate}%
Final: ${equity[-1]:.0f}
Return: {ret:.1%}
Max DD: {mdd:.1f}%
------------------------
Short Supp: ${ss}
Short Res: ${sr}
Long Supp: ${ls}
Long Res: ${lr}
------------------------
Trend: {trend[:18]}
Flow: {flow[:22]}
"""
    ax3.text(0.05,0.95,txt,va='top',fontsize=10,color='white',
             bbox=dict(facecolor='#222',edgecolor='gray',pad=10))

    fig.suptitle(f"{name} Backtest",fontsize=16,c='white',weight='bold')
    plt.tight_layout()
    plt.show()
    plt.style.use('default')

# ====================== 主分析函数（K线 + 期权 + 回测） ======================
def analyze_stock(ticker):
    real_code = auto_code(ticker)
    print("\n" + "="*60)
    print(f"📊 输入代码：{ticker}   👉   实际标的：{real_code}")
    print("="*60)

    # 1. 获取股价数据
    stock = yf.Ticker(real_code)
    df = stock.history(period="3mo")
    close_price = df["Close"].iloc[-1]
    ma5 = df["Close"].rolling(5).mean()
    ma20 = df["Close"].rolling(20).mean()
    
    current_ma5 = ma5.iloc[-1]
    current_ma20 = ma20.iloc[-1]

    # 计算支撑/压力位
    support1 = close_price * 0.98
    support2 = current_ma20 * 0.97
    resistance1 = close_price * 1.02
    resistance2 = close_price * 1.05

    # ====================== 【股价图】 ======================
    plt.figure(figsize=(14, 6))
    plt.plot(df["Close"], label="收盘价", linewidth=2.5, color="#1f77b4", zorder=3)
    plt.plot(ma5, label="MA5均线", linewidth=2, color="#ff7f0e", zorder=2)
    plt.plot(ma20, label="MA20均线", linewidth=2, color="#2ca02c", zorder=2)
    plt.axhline(support1, color="green", linestyle="--", linewidth=2, alpha=0.8, label="短期支撑", zorder=1)
    plt.axhline(resistance1, color="red", linestyle="--", linewidth=2, alpha=0.8, label="短期压力", zorder=1)
    
    plt.scatter(df.index[-1], close_price, color="blue", s=100, zorder=5, label=f"当前价: {close_price:.2f}")
    plt.text(df.index[-1], close_price + 2, f"当前价\n{close_price:.2f}", ha="center", va="bottom", fontsize=10, fontweight="bold", color="blue")
    plt.text(df.index[-1], support1, f"支撑\n{support1:.2f}", ha="center", va="top", fontsize=9, color="green")
    plt.text(df.index[-1], resistance1, f"压力\n{resistance1:.2f}", ha="center", va="bottom", fontsize=9, color="red")
    
    plt.title(f"{real_code} 股价走势（近3个月）| 关键点位标注", fontsize=14, fontweight="bold")
    plt.xlabel("日期", fontsize=12)
    plt.ylabel("价格", fontsize=12)
    plt.legend(fontsize=11, loc="upper right")
    plt.grid(alpha=0.3, linestyle="-")
    plt.tight_layout()
    plt.show()

    # ====================== 技术面文字分析 ======================
    print("\n📊 ===== 技术指标综合分析 =====")
    print(f"📌 当前价格：{close_price:.2f}")
    print(f"📅 5日均线（MA5）：{current_ma5:.2f}")
    print(f"📅 20日均线（MA20）：{current_ma20:.2f}")
    print(f"🛡️ 短期支撑位：{support1:.2f}")
    print(f"🛡️ 长期支撑位：{support2:.2f}")
    print(f"⚠️ 短期压力位：{resistance1:.2f}")
    print(f"⚠️ 长期压力位：{resistance2:.2f}")
    print("-"*50)

    if close_price > current_ma5 and close_price > current_ma20:
        trend = "🚀 多头趋势（站稳所有均线）"
    elif close_price > current_ma20:
        trend = "📈 偏多趋势（站稳20日线）"
    elif close_price < current_ma5 and close_price < current_ma20:
        trend = "💥 空头趋势（跌破所有均线）"
    else:
        trend = "🔍 震荡趋势"

    if close_price > resistance1:
        breakout = "✅ 已突破短期压力，继续看涨"
    elif close_price < support1:
        breakout = "❌ 已跌破短期支撑，注意风险"
    elif abs(close_price - resistance1) / close_price < 0.015:
        breakout = "⚡ 接近压力位，等待突破"
    elif abs(close_price - support1) / close_price < 0.015:
        breakout = "🛡️ 接近支撑位，等待反弹"
    else:
        breakout = "📌 处于区间中部"

    print(f"📶 趋势状态：{trend}")
    print(f"🔥 突破状态：{breakout}")
    print("="*60)

    # ====================== 期权数据分析 ======================
    print("\n🔥 期权深度数据")
    try:
        opt = stock.option_chain()
        calls = opt.calls
        puts = opt.puts

        print(f"✅ 期权数据获取成功")
        print(f"看涨期权(CALL)：{len(calls)} ｜ 看跌期权(PUT)：{len(puts)}")

        big_calls = calls[calls["volume"] >= 500].sort_values("volume", ascending=False)
        big_puts  = puts[puts["volume"] >= 500].sort_values("volume", ascending=False)
        print("\n===== 看涨期权大单（≥500）=====")
        print(big_calls[["strike","volume","openInterest","impliedVolatility"]].head(5))
        print("\n===== 看跌期权大单（≥500）=====")
        print(big_puts[["strike","volume","openInterest","impliedVolatility"]].head(5))

        cv = calls["volume"].sum()
        pv = puts["volume"].sum()
        print(f"\n【💸 资金流向】CALL：{cv:.0f} ｜ PUT：{pv:.0f} ｜ 多空比：{cv/(pv+0.001):.2f}")

        oi_c = calls["openInterest"].sum()
        oi_p = puts["openInterest"].sum()
        pcr = oi_p / (oi_c + 0.001)
        pcr_desc = "极端看多" if pcr < 0.5 else "偏多" if pcr < 0.7 else "中性" if pcr < 1 else "偏空" if pcr < 1.5 else "极端看空"
        print(f"【🧠 PCR情绪】{pcr:.2f} → {pcr_desc}")

        opt_all = pd.concat([calls, puts])
        valid_mask = abs(opt_all["strike"] - close_price) < close_price * 0.3
        opt_all_valid = opt_all[valid_mask]
        opt_all_valid["pain"] = opt_all_valid["openInterest"] * np.abs(opt_all_valid["strike"] - close_price)
        max_pain = opt_all_valid.groupby("strike")["pain"].sum().idxmin()
        print(f"【🎯 最大痛苦价】MaxPain = {max_pain:.2f}")

        iv_mask = abs(opt_all_valid["strike"] - close_price) < close_price * 0.2
        iv_calls = calls[calls["strike"].isin(opt_all_valid[iv_mask]["strike"])]
        iv_puts = puts[puts["strike"].isin(opt_all_valid[iv_mask]["strike"])]
        iv_calls_sorted = iv_calls.sort_values("strike")
        iv_puts_sorted = iv_puts.sort_values("strike")
        ivc = iv_calls_sorted["impliedVolatility"].mean()
        ivp = iv_puts_sorted["impliedVolatility"].mean()
        iv_avg = (ivc + ivp) / 2
        iv_level = "极高" if iv_avg > 0.8 else "偏高" if iv_avg > 0.5 else "中等" if iv_avg > 0.3 else "偏低"
        print(f"【📊 波动率】CALL-IV：{ivc:.1%} ｜ PUT-IV：{ivp:.1%} ｜ 等级：{iv_level}")

        plt.figure(figsize=(16, 8))
        plt.bar(calls["strike"], calls["openInterest"], color="#2ca02c", alpha=0.7, label="看涨期权(CALL)", zorder=2)
        plt.bar(puts["strike"], puts["openInterest"], color="#d62728", alpha=0.7, label="看跌期权(PUT)", zorder=2)
        
        plt.axvline(close_price, color="#1f77b4", linewidth=3, linestyle="-", label=f"当前价格: {close_price:.1f}", zorder=4)
        plt.axvline(max_pain, color="#ff7f0e", linewidth=3, linestyle="--", label=f"MaxPain: {max_pain:.1f}", zorder=3)
        strong_resistance = close_price * 1.15
        strong_support = close_price * 0.98
        plt.axvline(strong_resistance, color="green", linewidth=2, linestyle="--", alpha=0.8, label=f"强压力: {strong_resistance:.1f}", zorder=3)
        plt.axvline(strong_support, color="red", linewidth=2, linestyle="--", alpha=0.8, label=f"强支撑: {strong_support:.1f}", zorder=3)

        top3_calls = big_calls.head(3)
        for idx, (_, row) in enumerate(top3_calls.iterrows()):
            plt.scatter(row["strike"], row["openInterest"], color="green", s=120, marker="*", edgecolor="black", zorder=5, label="认购异动" if idx == 0 else "")
            plt.text(row["strike"], row["openInterest"] + 500, f"{row['strike']:.0f}\n{row['volume']:.0f}", ha="center", va="bottom", fontsize=9, fontweight="bold", color="green")
        top3_puts = big_puts.head(3)
        for idx, (_, row) in enumerate(top3_puts.iterrows()):
            plt.scatter(row["strike"], row["openInterest"], color="red", s=120, marker="*", edgecolor="black", zorder=5, label="认沽异动" if idx == 0 else "")
            plt.text(row["strike"], row["openInterest"] + 500, f"{row['strike']:.0f}\n{row['volume']:.0f}", ha="center", va="bottom", fontsize=9, fontweight="bold", color="red")

        plt.axvspan(close_price * 0.9, close_price * 1.1, color="yellow", alpha=0.1, zorder=1, label="平值期权核心区")
        plt.text(close_price, plt.ylim()[1] * 0.95, "平值核心区\n(多空博弈关键)", ha="center", va="top", fontsize=11, fontweight="bold", bbox=dict(facecolor="yellow", alpha=0.5))

        plt.title(f"{real_code} 期权未平仓合约(OI)分布 | 关键点位+异动标注", fontsize=16, fontweight="bold")
        plt.xlabel("行权价", fontsize=13)
        plt.ylabel("未平仓合约数(OI)", fontsize=13)
        plt.legend(fontsize=12, loc="upper right")
        plt.grid(alpha=0.3, linestyle="-")
        plt.xlim(close_price * 0.5, close_price * 1.5)
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(16, 6))
        plt.plot(iv_calls_sorted["strike"], iv_calls_sorted["impliedVolatility"], color="green", linewidth=2.5, label="CALL-IV", zorder=3)
        plt.plot(iv_puts_sorted["strike"], iv_puts_sorted["impliedVolatility"], color="red", linewidth=2.5, label="PUT-IV", zorder=3)
        
        plt.scatter(close_price, iv_avg, color="blue", s=100, zorder=5, label=f"平值IV: {iv_avg:.1%}")
        plt.text(close_price, iv_avg + 0.05, f"平值IV\n{iv_avg:.1%}", ha="center", va="bottom", fontsize=10, fontweight="bold", color="blue")
        plt.text(plt.xlim()[0], plt.ylim()[1] * 0.9, "IV微笑曲线: 虚值期权波动率更高", fontsize=11, fontweight="bold", color="black")

        plt.title(f"{real_code} 期权波动率曲线(IV) | 平值标注", fontsize=14, fontweight="bold")
        plt.xlabel("行权价", fontsize=12)
        plt.ylabel("隐含波动率(IV)", fontsize=12)
        plt.legend(fontsize=11, loc="upper right")
        plt.grid(alpha=0.3, linestyle="-")
        plt.xlim(close_price * 0.7, close_price * 1.3)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"⚠️ 期权数据获取失败：{e}")

    # ====================== 策略回测 ======================
    print("\n" + "="*60)
    print("📈 开始策略回测（长期数据）")
    print("="*60)
    
    df_long = yf.download(real_code, start="2023-01-01", progress=False)
    close_list = to_list(df_long["Close"])
    ss, sr, ls, lr = support_resistance(close_list)
    trend = trend_direction(close_list)
    flow = option_flow_estimate(close_list)

    bh = backtest_bh(df_long)
    macd = backtest_macd(df_long)
    rsi = backtest_rsi(df_long)
    env = backtest_ma_envelope(df_long)
    osc = backtest_ma_osc(df_long)

    strategy_map = {"bh": bh, "macd": macd, "rsi": rsi, "env": env, "osc": osc}
    all_strats = [("Buy & Hold", *bh), ("MACD", *macd), ("RSI", *rsi), ("MA Envelope", *env), ("MA Oscillator", *osc)]

    print("="*85)
    print(f"{'Strategy':<16} {'Final':<12} {'Return':<10} {'Max DD':<10} {'Win Rate':<10}")
    print("="*85)
    for name, val, ret, trades, ind, ind_name, equity, mdd, wr in all_strats:
        print(f"{name:<16} ${val:<10.0f} {ret:<9.1%} {mdd:<8.1f}% {wr:<8.1f}%")
    print("="*85)

    best_return = -999
    best_data = None
    for s in all_strats:
        if s[2] > best_return:
            best_return = s[2]
            best_data = s

        # ====================== 自动生成 5 张策略图 ======================
    print("\n🚀 正在生成 5 张策略回测图，请稍候...")
    
    # 逐个绘制所有 5 个策略
    strategy_names = ["Buy & Hold", "MACD", "RSI", "MA Envelope", "MA Oscillator"]
    for i, s in enumerate([bh, macd, rsi, env, osc]):
        name = strategy_names[i]
        val, ret, trades, ind, ind_name, equity, mdd, wr = s
        
        plot_bloomberg(
            df_long, name, trades, ind, ind_name, equity,
            ret, mdd, wr, (ss,sr,ls,lr), trend, flow
        )
    
        # ====================== 【具体买卖点位建议】 ======================
        print(f"\n===== {name} 策略 · 具体买卖规则 & 点位建议 =====")

        if name == "Buy & Hold":
            print(f"• 策略：中长期价值持有，结合关键支撑压力布局")
            print(f"• 入场区间：靠近短期支撑 {ss:.2f} ~ 长期支撑 {ls:.2f} 分批建仓")
            print(f"• 加仓点：回踩 {ss:.2f} 附近企稳可加")
            print(f"• 防守线：有效跌破 {ls:.2f} 减仓/止损（防止趋势走坏）")
            print(f"• 目标区域：上方压力 {sr:.2f} ~ {lr:.2f} 可分批减仓锁定利润")

        elif name == "MACD":
            print(f"• 入场：MACD 线上穿信号线（金叉）→ 做多")
            print(f"• 止损：跌破短期支撑 {ss:.2f} 或 浮亏5% 离场")
            print(f"• 止盈：MACD 下穿信号线（死叉）→ 卖出")

        elif name == "RSI":
            print(f"• 入场：RSI 跌破 30（超卖）→ 分批低吸")
            print(f"• 止损：跌破短期支撑 {ss:.2f} 离场")
            print(f"• 止盈：RSI 升破 70（超买）→ 分批止盈")

        elif name == "MA Envelope":
            print(f"• 入场：价格回踩靠近短期支撑 {ss:.2f} 做多")
            print(f"• 止损：跌破 {ss:.2f} 止损")
            print(f"• 止盈：反弹至短期压力 {sr:.2f} 卖出")

        elif name == "MA Oscillator":
            print(f"• 入场：MA10 上穿 MA30 + 站稳 {ss:.2f} 上方")
            print(f"• 止损：跌破 {ss:.2f} 离场")
            print(f"• 止盈：接近压力位 {sr:.2f} 或 MA10 下穿 MA30 离场")

        # 统一点位
        print(f"\n📌 当前关键点位参考：")
        print(f"• 短期支撑：{ss:.2f}   短期压力：{sr:.2f}")
        print(f"• 长期支撑：{ls:.2f}   长期压力：{lr:.2f}")
        print("="*60)

# ====================== 启动程序 ======================
if __name__ == "__main__":
    analyze_stock(input("请输入股票代码：").strip())



# ================================================================================
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------- 3个市场各50只代表性股票 ----------------------
us_50 = [
    "AAPL", "MSFT", "GOOG", "AMZN", "NVDA", "TSLA", "META", "AVGO", "V", "MA",
    "JPM", "BAC", "WMT", "DIS", "NFLX", "PEP", "KO", "MRK", "JNJ", "PFE",
    "HD", "MCD", "IBM", "INTC", "CSCO", "ORCL", "CRM", "ADBE", "NKE", "T",
    "XOM", "CVX", "BA", "CAT", "MMM", "GE", "UNH", "HON", "QCOM", "RTX",
    "SPGI", "BLK", "AXP", "GS", "MS", "C", "AMGN", "GILD", "ISRG", "TMO"
]

hk_50 = [
    "0700.HK", "0941.HK", "0005.HK", "0003.HK", "0012.HK", "0883.HK", "0388.HK", "0939.HK", "1299.HK", "0066.HK",
    "0027.HK", "0960.HK", "0992.HK", "0981.HK", "shturl.", "0017.HK", "0083.HK", "0101.HK", "0151.HK", "0291.HK",
    "0688.HK", "0968.HK", "0268.HK", "0358.HK", "0175.HK", "0241.HK", "shturl.", "0386.HK", "0857.HK", "0014.HK",
    "0001.HK", "shturl.", "0004.HK", "0019.HK", "0023.HK", "0144.HK", "0152.HK", "0257.HK", "0293.HK", "0330.HK",
    "0380.HK", "0669.HK", "0762.HK", "0823.HK", "1038.HK", "1044.HK", "1093.HK", "1137.HK", "1810.HK"
]

cn_50 = [
    "600036.SS", "601318.SS", "600030.SS", "601166.SS", "601398.SS", "601988.SS", "600000.SS", "600104.SS", "600016.SS", "600887.SS",
    "000858.SZ", "000568.SZ", "002594.SZ", "002415.SZ", "000725.SZ", "002230.SZ", "002555.SZ", "002739.SZ", "002456.SZ", "002241.SZ",
    "600519.SS", "600809.SS", "601899.SS", "601888.SS", "600188.SS", "601088.SS", "601288.SS", "601998.SS", "601668.SS", "601601.SS",
    "600048.SS", "600309.SS", "600900.SS", "600028.SS", "601727.SS", "601360.SS", "600703.SS", "600089.SS", "600177.SS", "600276.SS",
    "000333.SZ", "000001.SZ", "000063.SZ", "000157.SZ", "000408.SZ", "000776.SZ", "000938.SZ", "002027.SZ", "002049.SZ", "002129.SZ"
]

market_stocks = {
    "US Market": us_50,
    "HK Market": hk_50,
    "CN A-Share": cn_50
}

# ---------------------- 批量回测主函数 ----------------------
def run_cross_market_backtest():
    all_results = []
    print("\n" + "="*80)
    print(" Starting Cross-Market Backtest: US / HK / CN (50 stocks each) ")
    print("="*80)

    for market_name, tickers in market_stocks.items():
        print(f"\n📊 Testing: {market_name} ...")

        for code in tickers:
            try:
                df = yf.download(code, start="2023-01-01", progress=False)
                if len(df) < 120:
                    continue

                # 直接调用你原来的5个策略
                bh_val, bh_ret, _, _, _, _, _, _ = backtest_bh(df)
                ma_val, ma_ret, _, _, _, _, _, _ = backtest_macd(df)
                rs_val, rs_ret, _, _, _, _, _, _ = backtest_rsi(df)
                env_val, env_ret, _, _, _, _, _, _ = backtest_ma_envelope(df)
                osc_val, osc_ret, _, _, _, _, _, _ = backtest_ma_osc(df)

                all_results.append({
                    "market": market_name,
                    "code": code,
                    "BuyHold": bh_ret,
                    "MACD": ma_ret,
                    "RSI": rs_ret,
                    "MAEnv": env_ret,
                    "MAOsc": osc_ret
                })
            except:
                continue

    # 汇总平均收益
    df_result = pd.DataFrame(all_results)
    summary = df_result.groupby("market")[["BuyHold", "MACD", "RSI", "MAEnv", "MAOsc"]].mean()
    print("\n" + "="*80)
    print(" Cross-Market Strategy Performance (Average Return) ")
    print("="*80)
    print(summary.round(2))

    # 绘图
    plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
    plt.figure(figsize=(14, 6))
    summary.T.plot(kind='bar', ax=plt.gca(), width=0.75)
    plt.title("Average Return of 5 Strategies Across 3 Markets (50 Stocks Each)", fontsize=14)
    plt.ylabel("Average Total Return (%)", fontsize=12)
    plt.xlabel("Strategy", fontsize=12)
    plt.grid(alpha=0.3)
    plt.legend(title="Market")
    plt.tight_layout()
    plt.show()

    return summary

# ---------------------- 运行！----------------------
if __name__ == "__main__":
    run_cross_market_backtest()
```
